```python
a = np.arange(0, 12)
A = np.reshape(a, (4, 3))
At = A.T
B = np.dot(A, At)

import numpy.linalg as lg

a, A, At, B, lg.det(B)
# Матрица B размера 4x4. Обратную вычислить нельзя, так как определитель равен 0 (в силу того, что определители перемножаемых матрицы равны 0, так как есть линейно зависимые строки).
```

```python
np.random.seed(42)
```

```python
c = (16 - 0) * np.random.random_sample(16) + 0
c
```

```python
C = np.reshape(c, (4, 4))
C
```

```python
D = 10 * C + B
D_inv = lg.inv(D) 

D, lg.det(D), lg.matrix_rank(D), D_inv
```

```python
D_inv_old = D_inv

D_inv = np.full_like(D_inv_old, 0)

D_inv[D_inv_old < 0] = 0
D_inv[D_inv_old > 0] = 1

D_inv
```

```python
E = np.where(D_inv, B, C)
E
```
